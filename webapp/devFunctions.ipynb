```python
import pandas as pd
import youtube_dlc

from stack.Stack import stackScrape
```

```python
stack = stackScrape()
```

```python
params = {	
            'base_url': 'https://stats.stackexchange.com/questions/tagged/',
            'tag': 'python',
            'query_filter': 'Votes',
            'max_pages': 50,
            'pagesize': 50
        }
```

```python
dfStack = stack.TagsStack(**params)
```

```python
dfStack['Tag'].to_list()
```

```python
def getData(base_url, tag, query_filter, max_pages, pagesize):
    
    ydl = youtube_dlc.YoutubeDL({'ignoreerrors':True})
    dfStack = stack.TagsStack(base_url, tag, query_filter, max_pages, pagesize)
    arrayQuery = dfStack['Tag'].to_list()
    
    results = list()
    for query in arrayQuery:
        queryConcat = query + ' python'
        r = ydl.extract_info('ytsearchdate1:\{}'.format(queryConcat), download=False)
        for entry in r['entries']:
            if entry is not None:
                entry['query'] = query
        results += r['entries']
    results = [e for e in results if e is not None]
    df = pd.DataFrame(results)
    
    return df
```

```python
df = getData(**params)
```
