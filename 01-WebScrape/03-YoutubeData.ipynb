```python
results = list()
for query in array_querry:
    query_concat = query + ' python'
    r = ydl.extract_info('ytsearchdate2:\{}'.format(query_concat), download=False)
    for entry in r['entries']:
        if entry is not None:
            entry['query'] = query
    results += r['entries']
```

```python
results = [e for e in results if e is not None]
len(results)
```

```python
df = pd.DataFrame(results)
df['query'] = "machine learning"

#columns = ['title', 'upload_date', 'view_count']
df['upload_date'] = pd.to_datetime(df['upload_date'])
df['days_since_publication'] = (pd.to_datetime('2021-01-01') - df['upload_date']) / np.timedelta64(1,'D')
```
